```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as consts
from matplotlib.ticker import FormatStrFormatter
```

```python
R = 2
Rcrit = 9.4
v = 30 * 10**3
e0 = 1
w0 = 235.7*10**3
```

```python
L = Rcrit / (2 * w0)
C = 1 / (L * w0**2)
wmax = 1 / np.sqrt(L*C)
print(L)
print(C)
print(wmax * 10**-3)
```

```python
plt.figure(figsize = (9,9))
plt.ylabel("$I(A)$")
plt.rc('font', size=15)
plt.xlabel("$w(КГц)$")
plt.grid();


w = np.arange(0.1, 10**3,1)
I = e0 / (R**2 + (w * 10**3 *L-1/(C*w*10**3))**2)

#plt.yticks([0, 10, pmin, 20, 30, 40, 50])
plt.xticks([0, 235.7, 400, 600, 800, 1000])
plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.1f'))

plt.plot(w, I, '-', linewidth = 2, label = r"$I(w)$")
plt.legend(loc="best");
plt.savepic("I(w).svg")
```
